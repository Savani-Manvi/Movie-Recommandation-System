```python
#Import library
import pandas as pd
```

```python
#Load data
ratings = pd.read_csv("../outputs/ratings_cleaned.csv")
movies = pd.read_csv("../outputs/movies_cleaned.csv")
```

```python
#Make new dataframe
movie_ratings_full = ratings.merge(movies, on="movieId")
```

```python
#Create movie summary table
movie_summary = ratings.groupby("movieId").agg(
    rating_count=("rating", "count"),
    average_rating=("rating", "mean")
).reset_index()
```

```python
#Merge with movies
movie_summary = movie_summary.merge(movies, on="movieId")
```

```python
#Genre
genre_data = movies.assign(
    genre=movies["genres"].str.split("|")
).explode("genre")
```

```python
#Save final Power BI dataset
movie_ratings_full.to_csv("../tableau/movie_ratings_full.csv", index=False)
movie_summary.to_csv("../tableau/movie_summary.csv", index=False)
genre_data.to_csv("../tableau/genre_data.csv", index=False)
```
