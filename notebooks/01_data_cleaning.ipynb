```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Load datasets
movies = pd.read_csv("../data/movies.csv")
ratings = pd.read_csv("../data/ratings.csv")
tags = pd.read_csv("../data/tags.csv")
links = pd.read_csv("../data/links.csv")
```

```python
#Check first rows
print("Movies\n", movies.head())
print("Ratings\n", ratings.head())
```

```python
#Check dataset shape
print("Movies\n", movies.shape)
print("Ratings\n", ratings.shape)
```

```python
#Check missing values
print("Movies\n", movies.isnull().sum())
print("Ratings\n", ratings.isnull().sum())
```

```python
#Check duplicate rows
print("Movies\n", movies.duplicated().sum())
print("Ratings\n", ratings.duplicated().sum())
```

```python
#Remove duplicate if any
movies = movies.drop_duplicates()
ratings = ratings.drop_duplicates()
```

```python
#Extract movie year from title
movies["year"] = movies["title"].str.extract(r"\((\d{4})\)")
```

```python
#Clean movie title
movies["clean_title"] = movies["title"].str.replace(r"\(\d{4}\)", "", regex=True).str.strip()
```

```python
#Save cleaned data
movies.to_csv("../outputs/movies_cleaned.csv", index=False)
ratings.to_csv("../outputs/ratings_cleaned.csv", index=False)
```
