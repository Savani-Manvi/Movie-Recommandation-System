```python
#Import library
import pandas as pd
```

```python
#Load data
ratings = pd.read_csv("../outputs/ratings_cleaned.csv")
movies = pd.read_csv("../outputs/movies_cleaned.csv")
```

```python
#Create movie summary table
movie_summary = ratings.groupby("movieId").agg(
    rating_count=("rating", "count"),
    average_rating=("rating", "mean")
).reset_index()
```

```python
#Merge with movies
movie_summary = movie_summary.merge(movies, on="movieId")
```

```python
#Mearge with rating
movie_ratings = ratings.merge(movies, on="movieId")
```

```python
#Genre
genre_data = movies.assign(
    genres=movies["genres"].str.split("|")
).explode("genres")
```

```python
#Save final Power BI dataset
movie_summary.to_csv("../powerbi/movie_summary.csv", index=False)
movie_ratings.to_csv("../powerbi/movie_ratings.csv", index=False)
genre_data.to_csv("../powerbi/genre_data.csv", index=False)
```
