```python
#Import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Load movies
movies = pd.read_csv("../outputs/movies_cleaned.csv")
```

```python
#Prepare genre text
movies["genres_text"] = movies["genres"].str.replace("|", " ", regex=False)
```

```python
#Convert genres into vectors
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(movies["genres_text"])
```

```python
#Calculate similarity
similarity = cosine_similarity(tfidf_matrix)
```

```python
#Create recommendation function
def recommend_movie(movie_name, total_recommendations=10):
    movie_index = movies[movies["clean_title"].str.lower() == movie_name.lower()].index

    if len(movie_index) == 0:
        return "Movie not found"

    movie_index = movie_index[0]
    similarity_scores = list(enumerate(similarity[movie_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    recommended_indexes = [i[0] for i in similarity_scores[1:total_recommendations+1]]

    return movies.iloc[recommended_indexes][["movieId", "clean_title", "genres", "year"]]
```

```python
#Test recommendation
recommend_movie("Toy Story")
```

```python
#Save sample recommendations
sample_recommendations = recommend_movie("Toy Story")
sample_recommendations.to_csv("../outputs/content_based_recommendations.csv", index=False)
```
